```python
import numpy as np
import plotly.graph_objects as go

# Define the function f(x1, x2) = 2 - x1^2 - x2^2
def func(x1, x2):
    return 2 - x1**2 - x2**2

# Create a grid of values
x1 = np.linspace(-2, 2, 50)
x2 = np.linspace(-2, 2, 50)
X1, X2 = np.meshgrid(x1, x2)
Z = func(X1, X2)

# Create the 3D surface plot
surface = go.Surface(z=Z, x=X1, y=X2, colorscale='viridis', opacity=0.8)

# Outline the base shape in the 2D plane
theta = np.linspace(0, 2*np.pi, 100)
circle_x = np.cos(theta) * np.sqrt(2)  # x1 values for base outline
circle_y = np.sin(theta) * np.sqrt(2)  # x2 values for base outline
circle_z = np.full_like(circle_x, -1)  # Base lies at z=-1

base_outline = go.Scatter3d(x=circle_x, y=circle_y, z=circle_z,
                            mode='lines', line=dict(color='red', width=3),
                            name="Base Outline")

# Define layout for interactivity
layout = go.Layout(
    title="Interactive 3D Plot of f(x1, x2) = 2 - x1 - x2",
    scene=dict(
        xaxis_title="x1",
        yaxis_title="x2",
        zaxis_title="f(x1, x2)",
        aspectmode="manual", aspectratio=dict(x=1, y=1, z=0.5),
    ),
    margin=dict(l=0, r=0, t=40, b=0),
)

# Create the figure and show it
fig = go.Figure(data=[surface, base_outline], layout=layout)
fig.show()
```
